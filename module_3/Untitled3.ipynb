```python
class DepartmentReport():
    def __init__(self, company_name):
        self.company_name = company_name
        
    def add_revenue(self,  revenue):
        if not hasattr(self, 'revenues'):  
            self.revenues= []  

        self.revenues.append(revenue)
     
    def average_revenue(self):
       return f"Average department revenue for {self.company_name}: {round(sum(self.revenues)/len(self.revenues))}"
    

    
```

```python
report = DepartmentReport("Danon")
report.add_revenue(1_000_000)
report.add_revenue(400_000)

print(report.average_revenue())
# => Average department revenue for Danon: 700000
      
```

```python
class User():
    def __init__(self, email, passsword, balance):
        self.email = email
        self.passsword = passsword
        self.balance = balance 
        
    def login(self, email, passsword):
        if self.email == email and self.passsword == passsword:
            return True
        else:
            False
    def update_balance(self, amount):
        self.balance+=amount
```

```python
class IntDataFrame():
    def __init__(self, nums):
        self.nums = [int(x) for x in nums]
    
    def count(self):
        self.pos_list = [x for x in self.nums if x!=0]
        return len(self.pos_list)
        
    def unique(self):
        return len(set(self.nums))
```

```python
df = IntDataFrame([4.7, 4, 3, 0, 2.4, 0.3, 4])

df.count()
# => 5

```

```python

class OwnLogger():
    def __init__(self):
        self.logs = []
                
    def log(self, message, level):
        self.logs.append([message, level])
    
    def show_last(self, level='all'):
        if level=='all':
            return self.logs[-1][0]
        else:
            for log in self.logs[::-1]:
                if log[1]==level:
                    return log[0]
            return None
```

```python
logger = OwnLogger()
logger.log("System started", "info")
logger.show_last("error")
# => None
# Некоторые интерпретаторы Python могут не выводить None, тогда в этой проверке у вас будет пустая строка
logger.log("Connection instable", "warning")
logger.log("Connection lost", "error")

logger.show_last()
# => Connection lost
logger.show_last("info")
# => System started
```
